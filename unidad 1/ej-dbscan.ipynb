Una de las cosas importantes sobre el método DBSCAN es que se deben elegir sus parámetros de manera cuidadosa. Una de las primeras impresiones es que si el radio de los círculos aumenta, el número de clusters disminuirá, ya que los clusters tendrán más puntos. Esto no siempre es cierto y se verá en este ejercicio.



````python
#| echo: True

import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=5000, 
                           n_features=10, 
                           n_redundant=0, 
                           n_informative=10,
                           random_state=1, 
                           n_clusters_per_class=1)
X.shape
````

Primero, con el conjunto de datos dado por el código anterior, seleccionar varios radios empezando en 2 y terminando en 2.8. Para cada radio $r$, usar la configuración 

```
db = DBSCAN(eps=r, min_samples=5)
```

y graficar el número de clusters en función del radio. Se espera que la salida sea una gráfica donde disminuya el número de clusters mientras aumenta el radio.


````python
R = []
C = []
for r in range(200,280,5):
    r = r / 100
    R.append(r)
    db = DBSCAN(eps=r, min_samples=5)
    db.fit(X)
    y_c = db.labels_
    n_clusters = len(set(y_c)) - (1 if -1 in y_c else 0)
    C.append(n_clusters)
    #print(f"Eps: {r} clusters:{n_clusters}")
plt.plot(R,C)
plt.xlabel("Radio")
plt.ylabel("# clusters")
plt.show()
````

Luego, hacer lo mismo pero para radios en el rango 1.5 a 2. En este caso se debe ver el comportmamiento inverso.

````python
R = []
C = []
for r in range(150,200,5):
    r = r / 100
    R.append(r)
    db = DBSCAN(eps=r, min_samples=5)
    db.fit(X)
    y_c = db.labels_
    n_clusters = len(set(y_c)) - (1 if -1 in y_c else 0)
    C.append(n_clusters)
    #print(f"Eps: {r} clusters:{n_clusters}")
plt.plot(R,C)
plt.xlabel("Radio")
plt.ylabel("# clusters")
plt.show()
````
