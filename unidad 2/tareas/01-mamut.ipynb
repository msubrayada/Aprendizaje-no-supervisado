```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
import umap
from sklearn import decomposition
from sklearn.manifold import TSNE
```

```python
X = np.load("datos.npy")
```

```python
km = KMeans(8)
y_km = km.fit_predict(X)
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1,projection='3d')
ax.scatter(X[:,0], X[:,1], X[:,2], c = y_km, cmap='magma')
ax.set_title("Puntos del mamut")
plt.show()
```

# PCA

```python
pca = decomposition.PCA(n_components=2)
pca.fit(X)
X_t = pca.transform(X)
plt.scatter(X_t[:, 0],
            X_t[:, 1],
            c=y_km,
            s=0.5,
            cmap='magma')
plt.title("PCA - Puntos del mamut")
plt.show()
```

# T-SNE

```python
tsne = TSNE(n_components=2, perplexity=50)
X_tsne = tsne.fit_transform(X)
plt.scatter(X_tsne[:, 0],
            X_tsne[:, 1],
            c=y_km,
            s=0.5,
            cmap='magma')
plt.title("TSNE - Puntos del mamut")
plt.show()
```

# UMAP

```python
reducer = umap.UMAP(n_components=2, n_neighbors=10, min_dist=.2)
embedding = reducer.fit_transform(X)
```

```python
plt.scatter(embedding[:, 0],
            embedding[:, 1],
            c=y_km,
            s=0.5,
            cmap='magma')
plt.title("UMAP - Puntos del mamut")
plt.show()
```
